```python
# Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_data = pd.read_csv("/Users/gauravvraii/Documents/GitHub/Lab5100-Assignment8/venv/titanic_data.csv")
df_data.info
```

```python
df_data.head(10)
```

```python
df_data = df_data[['Survived','Pclass_1','Pclass_2','Pclass_3','Sex_male','Sex_female','Age','Fare']]
df_data.head()
```

# Cleaning Dataset

```python
#checking the missing values in the dataset
df_data.isnull().sum()
```

```python
#Grouping sex and pclass
Sex = df_data.groupby('Sex_male')['Sex_female'].mean()
Pclass = df_data.groupby('Pclass_1')['Pclass_2'].mean()
```

```python
df_data['Sex']=df_data.groupby("Sex_male")['Sex_female'].transform(lambda x: x.fillna(x.mean()))
df_data['Pclass']=df_data.groupby("Pclass_1")['Pclass_2'].transform(lambda x: x.fillna(x.mean()))
```

```python
df_data.tail(5)
```

```python
fig, ax = plt.subplots()
scatter = ax.scatter(df_data['Age'], df_data['Fare'], c=df_data['Sex'], cmap='viridis', label=df_data['Sex'].unique())
ax.set_title('Titanic Dataset: Age vs Fare')
ax.set_xlabel('Age')
ax.set_ylabel('Fare')
colorbar = plt.colorbar(scatter, ax=ax, label='Sex')
ax.legend(['Male', 'Female'])
plt.show()
```

```python
# get columns to plot
columns = df_data.columns.drop(['Fare'])
# create x data
x_data = range(0, df_data.shape[0])
# create figure and axis
fig, ax = plt.subplots()
# plot each column
for column in columns:
    ax.plot(x_data, df_data[column])
# set title and legend
ax.set_title('Iris Dataset')
ax.legend()
```
